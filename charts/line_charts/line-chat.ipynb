```python
import pandas as pd

unrate = pd.read_csv('unrate.csv')

unrate['DATE'] = pd.to_datetime(unrate['DATE'])

unrate[0:5]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(unrate['DATE'], unrate['VALUE'])
plt.show()
```

```python
first_twelve = unrate[0:12]
plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")

plt.show()
```

```python

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
```

```python

fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
data_1948 = unrate[0:12]
data_1949 = unrate[12:24]

ax1.plot(data_1948['DATE'], data_1948['VALUE'])
ax2.plot(data_1949['DATE'], data_1949['VALUE'])
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))

ax1 = fig.add_subplot(5,1,1)
ax2 = fig.add_subplot(5,1,2)
ax3 = fig.add_subplot(5,1,3)
ax4 = fig.add_subplot(5,1,4)
ax5 = fig.add_subplot(5,1,5)

def plot_axis(ax, start, end):
    ax.plot(unrate[start:end]['DATE'], unrate[start:end]['VALUE'])
             
plot_axis(ax1, 0, 12)
plot_axis(ax2, 12, 24)
plot_axis(ax3, 24, 36) 
plot_axis(ax4, 36, 48)
plot_axis(ax5, 48, 60)

plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize=(6,3))

plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='red')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='blue')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))

def plot_years(start, end, c):
    plt.plot(unrate[start:end]['MONTH'], unrate[start:end]['VALUE'], c=c)

plot_years(0,12,'red')
plot_years(12,24,'blue')
plot_years(24,36,'green')
plot_years(36,48,'orange')
plot_years(48,60,'black')

plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
legends = ['1948', '1949', '1950', '1951', '1952']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=legends[i])
    
plt.legend(loc='upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Monthly, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
