```python
recent_grads_sort_pay = recent_grads.sort_values(by='Median')
recent_grads_sort_pay.head(2)
```

```python
recent_grads_sort_pay.tail(2)
```

```python
ax = recent_grads_sort_pay[:10].plot.bar(x='Major' ,y='ShareWomen')
ax.set_ylim(0,1)
ax.set_title("Loweset paying jobs share women")
```

```python
rows = recent_grads_sort_pay.shape[0]
ax = recent_grads_sort_pay[rows-10:rows].plot.bar(x='Major' ,y='ShareWomen')
ax.set_ylim(0,1)
ax.set_title("Highest paying jobs share women")
```

```python
ax = recent_grads_sort_pay[:10].plot.bar(x='Major' ,y='Unemployment_rate')
ax.set_ylim(0,1)
ax.set_title("Loweset paying jobs vs unemplyment rate")
```

```python
ax = recent_grads_sort_pay[rows-10:rows]['Unemployment_rate'].plot(kind='bar')
ax.set_ylim(0,1)
ax.set_title("Highest paying jobs vs unemplyment rate")
```
