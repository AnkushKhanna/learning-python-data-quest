```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(15, 18))

def create_graph(list_cats, index):
    for sp in range(0,len(list_cats)):
        ax = fig.add_subplot(6,3,(sp*3)+index)
        ax.plot(women_degrees['Year'], women_degrees[list_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[list_cats[sp]], c=cb_orange, label='Men', linewidth=3)
        for k,s in ax.spines.items():
            #Reson the lines are not showing on top, bottom, left and right
            s.set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_yticks([0,100])
        ax.set_title(list_cats[sp])
        #Reason why the ticks are not showing
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        if sp == 0:
            ax.text(2005, 87, 'Men')
            ax.text(2002, 8, 'Women')
        elif sp == len(list_cats)-1:
            ax.text(2005, 62, 'Men')
            ax.text(2001, 35, 'Women')
            ax.tick_params(labelbottom='on')
        

create_graph(stem_cats,1)
create_graph(lib_arts_cats,2)
create_graph(other_cats,3)
plt.savefig('gender_degrees.png')
plt.show()
```
