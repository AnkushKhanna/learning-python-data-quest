```python
import pandas as pd
jeopardy = pd.read_csv('jeopardy.csv')
jeopardy.head()
```

```python
jeopardy.columns
```

```python
dict_columns = {}
for c in jeopardy.columns:
    dict_columns[c] = c.strip() 
dict_columns
```

```python
jeopardy = jeopardy.rename(columns=dict_columns)
jeopardy.columns
```

```python
import re
def normalize_str(value):
    value = value.lower()
    return re.sub("[^A-Za-z0-9\s]", "", value)

jeopardy['clean_question'] = jeopardy['Question'].apply(normalize_str)
```

```python
jeopardy['clean_answer'] = jeopardy['Answer'].apply(normalize_str)
```

```python
jeopardy.head()
```

```python
def convert_dollor(value):
    if ~pd.isnull(value):
        value = re.sub("[^A-Za-z0-9\s]", "", value)
        try:
            return int(value)
        except ValueError:
            return 0

jeopardy['clean_value'] = jeopardy['Value'].apply(convert_dollor)
jeopardy.head()
```

```python
jeopardy['Air Date'] = pd.to_datetime(jeopardy['Air Date'])
jeopardy.head()
```

```python
def question_close_to_answer(row):
    split_answer = row['clean_answer'].split(' ')
    split_question = row['clean_question'].split(' ')
    if "the" in split_answer:
        split_answer.remove("the")
    match_count = 0
    if len(split_answer) == 0:
        return 0
    for a in split_answer:
        if a in split_question:
            match_count +=1
    return match_count / len(split_answer)

jeopardy['answer_in_question'] = jeopardy.apply(question_close_to_answer, axis=1)
jeopardy['answer_in_question'].head(5)
```

```python
jeopardy['answer_in_question'].mean()
```

Mean is so low, that is is clear that the finding answer from questions is next to impossible. Thus using this strategy might not be a good way to proceed.

```python
question_overlap = []
terms_used = set()
for i, row in jeopardy.iterrows():
        split_question = row["clean_question"].split(" ")
        split_question = [q for q in split_question if len(q) > 5]
        match_count = 0
        for word in split_question:
            if word in terms_used:
                match_count += 1
        for word in split_question:
            terms_used.add(word)
        if len(split_question) > 0:
            match_count /= len(split_question)
        question_overlap.append(match_count)
jeopardy["question_overlap"] = question_overlap

jeopardy["question_overlap"].mean()
```

```python
def low_high(row):
    if row['clean_value'] > 800:
        return 1
    else:
        return 0

jeopardy['high_value'] = jeopardy.apply(low_high, axis=1)
jeopardy.head()
```

```python
def assign_words(word):
    low_count = 0 
    high_count = 0
    for i, r in jeopardy.iterrows():
        if word in r['clean_question'].split(' '):
            if r['high_value'] == 1:
                high_count += 1
            else:
                low_count += 1
    return high_count, low_count

observed_expected = []
comparison_terms = list(terms_used)[0:5]

for t in comparison_terms:
    observed_expected.append(assign_words(t))
```
