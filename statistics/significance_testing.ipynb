```python
sampling_distribution = {}

for diff in mean_differences:
    count = sampling_distribution.get(diff, 0)
    sampling_distribution[diff] = count + 1

sampling_distribution
```

```python
frequencies = []
for k, v in sampling_distribution.items():
    if k > 2.52:
        frequencies.append(v)
    
p_value = np.sum(frequencies)/1000
p_value
```
