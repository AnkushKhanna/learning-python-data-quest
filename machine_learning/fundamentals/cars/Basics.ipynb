```python
def concat_dummies(col, prefix):
    dummies = pd.get_dummies(cars[col], prefix=prefix)
    return pd.concat([cars, dummies], axis=1)
```

```python
cols_categorial = ['make','fuel-type','aspiration','num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'num-of-cylinders', ' fuel-system']
for c in cols_categorial:
    cars = concat_dummies(c, c)
    cars = cars.drop([c], axis=1)
```

```python
numeric_cols = ['symboling', 'normalized-losses', 'wheel-base', 'length', 'height', 'width', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg']
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score
import numpy as np
def knn_train_test(cols, df, k):
    kfold = KFold(len(df), 5, shuffle=True)
    knn = KNeighborsRegressor(n_neighbors = k)
    scores = cross_val_score(knn, df[cols], df['price'], scoring='mean_squared_error', cv=kfold)
    rmses = [abs(x) ** (1/2) for x in scores]
    return np.mean(rmses)

dict_score = {}
ks = [1,3,5,7,9]
for c in numeric_cols:
    for k in ks:
        if dict_score.get(c) == None:
            dict_score[c] = {}
        dict_score[c][k] = knn_train_test([c], cars, k)

dict_score
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for k in ks:
    scores_k = []
    for key, value in dict_score.items():
        scores_k.append(value[k])
    fig, ax = plt.subplots()
    plt.hist(scores_k)
    plt.title(k)
    ax.set_xlim(3000, 12000)
    plt.show()
```

```python
dict_score_5 = {}
for k,v in dict_score.items():
    dict_score_5[k] = v[5]

dict_score_5
```

```python
sorted_features = sorted(dict_score_5, key=dict_score_5.get)
```

```python
dict_feature = {}
for i in [2,3,4,5]:
    features = sorted_features[0:i]
    rmse = knn_train_test(features, cars, 5)
    dict_feature[i] = rmse

dict_feature
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in [2,3,4]:
    features = sorted_features[0:i]
    lowest_k = 0
    lowest_rmse = float("inf")
    for k in list(range(1,26)):
        rmse = knn_train_test(features, cars, k)
        if rmse < lowest_rmse:
            lowest_rmse = rmse
            lowest_k = k
    print(i)
    print(lowest_k)            
    print(lowest_rmse)
```
